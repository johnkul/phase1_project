Data Cleaning

```python
# Remove dollar signs and commas, then convert to integers
df_tnmovie['production_budget'] = df_tnmovie['production_budget'].replace('[\$,]', '', regex=True).astype(int)
df_tnmovie['domestic_gross'] = df_tnmovie['domestic_gross'].replace('[\$,]', '', regex=True).astype(int)
df_tnmovie['worldwide_gross'] = df_tnmovie['worldwide_gross'].replace('[\$,]', '', regex=True).astype(float)

print(df_tnmovie.dtypes)
print(df_tnmovie.head())
```

```python
#Summary Statistics
print(df_tnmovie.describe())
```

```python
# Check for missing values
print(df_tnmovie.isnull().sum())
```

Data Visualization

```python
#Distribution of Production Budget
plt.figure(figsize=(20, 8))
sns.histplot(df_tnmovie['production_budget'], bins = 20, kde=True)
plt.title('Distribution of Production Budget')
plt.xlabel('Production Budget')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Domestic Gross
plt.figure(figsize=(10, 6))
sns.histplot(df_tnmovie['domestic_gross'], bins = 20, kde=True)
plt.title('Distribution of Domestic Gross')
plt.xlabel('Domestic Gross')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Worldwide Gross
plt.figure(figsize=(20, 6))
sns.histplot(df_tnmovie['worldwide_gross'], bins = 20, kde=True)
plt.title('Distribution of Worldwide Gross')
plt.xlabel('Worldwide Gross')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot of Production Budget vs. Worldwide Gross
plt.figure(figsize=(10, 6))
sns.scatterplot(x='production_budget', y='worldwide_gross', data=df_tnmovie)
plt.title('Production Budget vs. Worldwide Gross')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

Correlation Analysis

```python
# Correlation matrix
corr = df_tnmovie.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Top Performers

```python
#Top 5 Movies by Production Budget
top_budget = df_tnmovie.sort_values(by='production_budget', ascending=False).head(5)
print("Top 5 Movies by Production Budget:")
print(top_budget[['movie', 'production_budget']])

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_budget['movie'], top_budget['production_budget'], color='skyblue')
plt.xlabel('Production Budget (in hundred millions)')
plt.ylabel('Movie')
plt.title('Top 5 Movies by Production Budget')
plt.gca().invert_yaxis()  # Invert y-axis to display top-down
plt.show()
```

# rt.movie_info.tsv file Manipulation

```python
# Load the data
rt_movie = pd.read_csv('rt.movie_info.tsv', delimiter='\t', index_col = 0)
rt_movie.head()
```

```python
rt_movie.info()
```

Check for missing values

```python
rt_movie.isna().sum()
```

```python
# Drop rows where all values are NaN
rt_movie.dropna(how='all', inplace=True)
rt_movie.isna().sum()
```

Summary statistics for numerical columns

```python
print (rt_movie.describe()) #before the $ is removed.
```

```python
# Get summary statistics for numerical columns
print(rt_movie.describe())

# Convert 'box_office' to numeric after removing currency symbols
rt_movie['box_office'] = rt_movie['box_office'].replace('[\$,]', '', regex=True).astype(float)

# Summary statistics after conversion
print(rt_movie['box_office'].describe())
```

Explore categorical variables

```python
# Unique values in 'rating', 'genre', and 'studio' columns
print(rt_movie['rating'].unique())
print(rt_movie['genre'].unique())
print(rt_movie['studio'].unique())

# Frequency counts for categorical columns
print(rt_movie['rating'].value_counts())
print(rt_movie['genre'].value_counts())
print(rt_movie['studio'].value_counts())
```

Visualize the data for better insights

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Distribution of 'box_office'
plt.figure(figsize=(10, 6))
sns.histplot(rt_movie['box_office'], bins=20, kde=True)
plt.title('Distribution of Box Office Revenue')
plt.xlabel('Box Office Revenue ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Number of movies per rating
plt.figure(figsize=(10, 6))
sns.countplot(data=rt_movie, x='rating', order=rt_movie['rating'].value_counts().index)
plt.title('Number of Movies per Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box office revenue by genre (top 10 genres)
top_genres = rt_movie['genre'].value_counts().head(10).index
filtered_rt_movie = rt_movie[rt_movie['genre'].isin(top_genres)]

plt.figure(figsize=(14, 8))
sns.boxplot(data=filtered_rt_movie, x='genre', y='box_office')
plt.title('Box Office Revenue by Genre (Top 10 Genres)')
plt.xlabel('Genre')
plt.ylabel('Box Office Revenue ($)')
plt.xticks(rotation=45)
plt.show()
```

# rt.reviews File Manipulation

```python
rt_reviews = pd.read_csv('rt.reviews.tsv', delimiter='\t', encoding='latin1', index_col = 0)
rt_reviews.head(10)
```

Understand the Data

```python
# Check the structure of the dataframe
print(rt_reviews.info())
```

```python
# Get summary statistics of the DataFrame for the numeric column(s)
print(rt_reviews.describe())
```

```python
# Check for missing values
print(rt_reviews.isna().sum())
```

Data Cleaning

```python
# Drop rows with all NaN values
rt_reviews_cleaned = rt_reviews.dropna(how='all')

# Fill NaN values in 'rating' with a placeholder (e.g., 'Not Rated')
rt_reviews_cleaned['rating'].fillna('Not Rated', inplace=True)

# Fill NaN values in 'critic' and 'publisher' with 'Unknown'
rt_reviews_cleaned['critic'].fillna('Unknown', inplace=True)
rt_reviews_cleaned['publisher'].fillna('Unknown', inplace=True)

# Drop duplicates if any
rt_reviews_cleaned.drop_duplicates(inplace=True)

# Convert 'date' column to datetime
rt_reviews_cleaned['date'] = pd.to_datetime(rt_reviews_cleaned['date'])

print(rt_reviews_cleaned.info())
```

```python
print(rt_reviews_cleaned['rating'].isnull().sum())
```

```python
print(rt_reviews_cleaned.head())
```

Exploratory Data Analysis

```python
# Plot the distribution of ratings

# Check the columns of the DataFrame
print(rt_reviews_cleaned.columns)

# Check for missing values in the 'rating' column
print(rt_reviews_cleaned['rating'].isnull().sum())

# Inspect the first few rows of the DataFrame
print(rt_reviews_cleaned.head())

# Plot the distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=rt_reviews_cleaned, x='rating', order=rt_reviews_cleaned['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the count of fresh vs rotten reviews
plt.figure(figsize=(8, 5))
sns.countplot(data=rt_reviews_cleaned, x='fresh')
plt.title('Fresh vs. Rotten Reviews')
plt.xlabel('Review Type')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the number of reviews over time
plt.figure(figsize=(12, 6))
rt_reviews_cleaned.set_index('date').resample('M').size().plot()
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Plot the distribution of reviews by top critics and regular critics
plt.figure(figsize=(10, 6))
sns.countplot(data=rt_reviews_cleaned, x='top_critic')
plt.title('Distribution of Reviews by Top Critics vs. Regular Critics')
plt.xlabel('Top Critic (1=Yes, 0=No)')
plt.ylabel('Count')
plt.show()
```
